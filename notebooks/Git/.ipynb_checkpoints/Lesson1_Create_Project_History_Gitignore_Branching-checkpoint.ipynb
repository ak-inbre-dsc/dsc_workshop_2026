#### &nbsp;&nbsp; 1.2.2 &nbsp;&nbsp;Using the Git Sidebar


To get Git to track your work, you need to initialize your **Git Repository**
A Git repository is a **digital folder that tracks the entire history of your project**, allowing you to save "snapshots" of your work so you can revisit or restore any version at any time.

> We will do this by using the **Git side bar**. Follow these steps to open the side bar and **Initialize Your Git Repository**:
>1. Click the **Git Icon** (the diamond with branches). 
>2. Select: **Initialize a Repository**
>3. Click **Yes** to the question if you want to make the directory a Git Repo.

Git will now create a hidden directory named .git inside your `my_first_git_project` folder. This .git directory is where Git stores all of your project's history, configurations, and internal workings. You typically don't need to interact with the .git directory directly.

You will see several things appear in the Git side bar now:
- **Current Repository**: my_first_git_project - Your current working directory that you are tracking.
- **Current Branch**: master - the "branch" you are working on (we will cover branches later).
- **Changes | History**: tabs that show your work and the history of your version control.
- **Staged**: Staging area (which we cover below).
- **Changed**: Changes saved to a particular file.
- **Untracked**: Files that are not tracked by Git yet.

### &nbsp;&nbsp; 1.3. &nbsp;&nbsp; Exploring the <span style="color: blue;">Working Directory</span> - Where Your Files Live

Right now, your <span style="color: blue;">Working Directory</span> is the `my_first_git_project` folder. It's empty.
Let's create a new file in this directory. This file will be only in your <span style="color: blue;">Working Directory</span>. Git doesn't know about it yet.

Run the following cell to create a simple R file.

````python
!echo "# This is my first line of code" > my_code.R
````

**Look at the Sidebar now:**

- Click the **Git Icon** if you are not already there.
- You’ll see your `my_code.R` file listed under **Untracked**.
- Look for the **"U"**. This is Git’s way of saying this file is **Unmodified**. 

The file is unmodified because we have not saved our work yet so Git does not know we made any changes. This is the first state of a file in the Git workflow, sitting in the <span style="color: blue;">Working Directory</span>, but not yet part of Git's tracking system.

Let's make sure our new file contains the line we want it to.

````python
!cat my_code.R
````

Did it work?

As our file is not tracked now, Git is not tracking it. We need to "move" our file from Untracked to the Staged, also called the <span style="color: red;">Staging Area</span>.

### &nbsp;&nbsp; 1.4. &nbsp;&nbsp; The <span style="color: red;">Staging Area</span> - Preparing Your Changes

Now, let's move `my_code.R` from being an "untracked file" in our <span style="color: blue;">Working Directory</span> to the <span style="color: red;">Staging Area</span>. The staging area is a middle ground where you hand-pick and organize specific changes you’ve made before officially "saving" them to your project's history with a commit.

Before Git can "track" your work, you must **save your progress**. However, in Git, saving the file is only the first step. Follow the steps below to save your change and move it to the staging area:

- **Save**: This updates the `my_code.R` file in your `my_first_git_project` folder.

- **Stage (+)**: Hover your cursor over your file name - several icons will apear. Click the plus sign next to the file in the sidebar. When you **Stage (+)** a file, you're essentially telling Git, "Hey, I want to include the current version of this file in the snapshot I'm about to save."

You will now see in the left git panel that your `my_code.R` file has moved to the **Staged** tab. 

**Your file is now ready to commit!**

---

### &nbsp;&nbsp; 1.5. &nbsp;&nbsp; The <span style="color:darkorange ;">Git Repository</span> - Committing Your Changes 

A **Commit** is the act of taking a permanent "snapshot" of your staged changes, creating a labeled point in your project's history that you can always return to later.

Once you have saved and staged your file, you will commit your changes in two steps:

1. **Add a commit message in the `Summary` text window in your Git sidebar**. This message should be a concise and descriptive summary of what changes you made and why. Good commit messages are crucial for understanding your project's history later on. Since you just made your file, you could for example use: `Created my_code.R file`.
2. **Click COMMIT**. You’ve now created a permanent "Save Point" in your project's history!
   
> 

> Do you now see 0's for all three stages of the committing process (Staged, Changed, Untracked) in your git sidebar? This means that your changes has been successfully committed to your project's history.

You've just completed the fundamental Git workflow for tracking changes!

# Lesson 1 | Section 3: Ignoring files in commits

What if we have files that we do not want Git to track for us, like the `my_code-checkpoint.R` file or other intermediate files created during data analysis? Here you will learn how to stop Git from tracking absolutely everything you do.

### &nbsp;&nbsp; 3.1. &nbsp;&nbsp; Creating the Files We Want to Ignore

Let’s create a few dummy files. The first three will end up in the `my_first_git_project` folder, and the second two in the new `penguin_pics` folder. 

````python
!cd ~/my_first_git_project # make sure you are in the right directory
!mkdir -p penguin_pics
!touch flipper.png bill.png penguin_pics/Adelie.jpg penguin_pics/Gentoo.jpg
````

Putting these files under version control would be a waste of disk space as they are pictures and will not be edited themselves. What’s worse, having them all listed could distract us from changes that actually matter, so let’s tell Git to ignore them.

### &nbsp;&nbsp; 3.2. &nbsp;&nbsp; Creating Your .gitignore File

Now that you know how to commit, we can create a file called `.gitignore` to act as a filter. This file tells Git to ignore specific files or folders so they don't clutter your workspace. The `.gitignore` file is a simple text file. Run the code below to create this file. **you won't see the file appear in the folder as this is a *hidden file***

````python
# Create a .gitignore file to hide the digital clutter
!touch .gitignore
````

We will add the following text to this file. Line 1 tells Git to ignore all `.png` files in the working directory (`my_first_git_project`) and line 2 tells Git to ignore the whole `penguin_pics/` directory: 

````python
!echo "*.png" >> .gitignore
!echo "penguin_pics/" >> .gitignore

# Did it work?
!cat .gitignore
````

The asterisk `*` acts as a wildcard that matches any sequence of characters, meaning `*.png` tells Git to ignore any file or folder that ends in `.png`, and ignore the `penguin_pics` directory. (If any of these files were already being tracked, Git would continue to track them.) 

You will see that the new `.png` files and the whole `penguin_pics` folder has disappeared from your Git sidebar. The only thing Git notices now is the newly-created `.gitignore` file. 

You might think we wouldn’t want to track it, but everyone we’re sharing our repository with will probably want to ignore the same things that we’re ignoring. 

#### &nbsp;&nbsp;**Ignoring your checkpoint files** &nbsp;&nbsp;
We want to add another file to our `.gitignore file`. Remember the `my_code-checkpoint.R` intermediate file? Let's add it! 

As you can see, we are using the wildcard again. Here `*checkpoint*` tells Git to ignore any file or folder that contains the word "checkpoint" anywhere in its name. This will be useful if we add any other files that also create checkpoints.

````python
!echo *checkpoint* >> .gitignore
````

````python
!cat .gitignore
````

Git will now ignore all files containing the word `checkpoint`, and you will see the file has disappeared from the Git sidebar. 

### &nbsp;&nbsp; 3.3. &nbsp;&nbsp; Staging and Committing your `.gitignore` File

The last step is to stage and commit your `.gitignore` file. 
Move the currently untracked `.gitignore` file to the staging area and commit it using the following commit message: "Added files to ignore to the .gitignore."

Do you see your latest commit in the History tab?

# Lesson 1 | Section 4: Branching & Merging


Git allows us to create isolated lines of development called **branches**. This lets you work on new features or fixes without affecting the main codebase. Once a feature is complete, your changes can be integrated back into your `master` branch through **merging**.

In this exercise, you will:
1. Create a new `Feature` Branch.
2. Modify the R file on your new branch.
3. Commit your changes to the new branch.
4. Merging your `Feature` Branch with your `master` branch

**Branches** let you and your collaborators update existing code or add new code without fear of overwriting each others’ files. All commits live on a branch. The default branch in a repository is called the main/master branch.


Source: https://haydar-ai.medium.com/learning-how-to-git-merging-branches-and-resolving-conflict-61652834d4b0


Before we start, set your working directory:

````python
%cd ~/my_first_git_project
````

We're starting with your `my_first_git_project` and `my_code.R` file on the `master` branch. Let's create a new branch to work on a Feature. This keeps your changes separate from the `master` branch until they're ready.

We'll create a branch called `feature/add-theme`.

- Open the Git sidebar. You will see at the top that your **Current Branch** is `master`.
- Click the downward arrow on the right end of the bar, and you will see a menu with a blue **New Branch** button.
- Name your new branch `feature/add-theme`. Ignore the filter window. 
- You will now automatically be working from the new branch. 

### &nbsp;&nbsp; 4.2. &nbsp;&nbsp; Modify the R file on your new branch  

You can see in the first line that we are now on the newly created feature/add-theme branch. Now that we're on our new branch (feature/add-theme), we will modify `my_code.R`.

### &nbsp;&nbsp; 4.3. &nbsp;&nbsp; Commit Your Changes to Your New Branch

Below you will add your changed file to the Staging Area (`+`), commit it to your Git Repository (commit message: "Feature: Added a minimal theme to the scatter plot"), and see what changed in the History tab.

You can see under the History tab that your edited file was successfully committed to your new branch!

You are now working on the new `feature/add-theme` branch and **changes you made to your `my_code.R` file are not present in the `my_code.R` on the `master` branch**. This way, you can work on a file and commit as many changes as you want, without all these small commits cluttering your master branch history.  

### &nbsp;&nbsp; 4.4. &nbsp;&nbsp; Merging Your Feature Branch with your main branch

Once your commit to the `Feature Branch` is complete and verified, you can merge it back into the `master` branch so it is up to date. This integrates all the changes you made on your `feature/add-theme` branch into the `master` codebase.

**To do so, you do the following:**
1. Open the branch section using the downward arrow next to the Current Branch section.
2. Select the `master` branch. You will now have switched back to your `master` branch.
3. Now, when you hover over your `feature/add-theme` branch, you will see a little icon on the right that looks like 3 dots connected with lines.
4. Click on the icon, which will merge your new branch into the `master` branch and incorporate the changes.

If you now open your `my_code.R` file now on the `master` branch you will see the changes you made!

### &nbsp;&nbsp; 4.5. &nbsp;&nbsp; Deleting Your Feature Branch  

Now that you have merged the changes you made to my_code.R with your master branch, it is time to get rid of the new feature branch. It's generally a recommended best practice to delete a feature branch after it has been successfully merged into main. 

Deleting a branch after merging keeps your repository clean and organized, preventing clutter. It **does not lose any history**, as all commits are preserved in the main branch's log. This practice also prevents confusion or accidental work on old branches, clearly signaling that the feature is complete and integrated into the main codebase, aligning with standard Git workflows.

For every new feature, bug fix, or independent piece of work, you should create a new branch from main. 

Once the work is successfully merged, the feature branch is no longer needed locally.
1. Open the branch list in the Git sidebar again.
2. Right-click the feature branch (`feature/add-theme`).
3. Select **Delete Branch** to keep your workspace organized.

**To delete your feature branch, click the little trashcan icon next to the merging icon.**

### *Additional Information - Running Git in Your Terminal*

You can also run all these steps using command line in your terminal. Below are the commands, what they do, and their equivalent action you performed using the Git sidebar. 

| Command | Explanation | JupyterLab Git Sidebar Action |
| :--- | :--- | :--- |
| `git status` | Shows the current state of files. | Automatically updated in the **Changes** section. |
| `git add <file>` | Stages specific changes. | Click the **+** (plus icon) next to an individual file. |
| `git add .` | Stages all changes. | Click the **+** (plus icon) next to the **Changed** header. |
| `git commit -m "commit message"` | Saves the staged snapshot. | Type a message in the box at the bottom and click **Commit**. |
| `git log --oneline` | Displays commit history. | Click the **History** tab at the top of the Git panel. |
| `git checkout main` | Switches your working directory to the **main** (or master) branch. | Click the **Branch Name** at the top of the sidebar and select "main" from the list. |
| `git pull` | Fetches changes from the remote server and merges them into your current branch. | Click the **Cloud Pull icon** (down arrow) at the top of the Git panel. |
| `git checkout -b <name>` | Creates a new branch and immediately switches you to it. | Click the **Branch Name** → Click **New Branch** → Enter name and click **Create Branch**. |
| `git merge <branch>` | Combines the history and changes of the specified branch into your current branch. | While on *main*, click the **Branch menu** (three dots or branch list) and select **Merge** for your feature branch. |
| `git branch -d <name>` | Deletes the specified branch locally (assuming it has been merged). | Right-click the branch name in the branch list and select **Delete**. |


### Summary: A Typical Workflow Using the Git Sidebar in JupyterLab

1. Initialize a Local Repository:
   Open the **Git sidebar** (branch icon). If the folder is not yet tracked, click the **Initialize a Repository** button to create your local repo.

2. Configure .gitignore:
   Create a new file named `.gitignore` (using for example `!touch .gitignore`). List files or folders you want Git to ignore (e.g., `.ipynb_checkpoints/`), then save it. This prevents "junk" files from appearing in your Git panel.

3. Start on master branch:  
   Open the **Git sidebar** (branch icon). Make sure you are on the `master` branch (shown at the top). Click the **Pull** button to update it.

4. Create a new branch:  
   Click the current branch name → select **“New Branch”** → enter a name like `feature/new-feature-name` (or `bugfix/bug-description`) → create and switch to it.

5. Work on the new branch:  
   Make your changes. 

6. Stage files:
   In the Git panel, find your modified files under the "Changed" or "Untracked" sections. Click the **“+”** button next to the files you want to include in your next save. They will move to the **Staged** section.

7. Commit:
   Once files are staged, add a commit message in the box at the bottom of the Git panel and click **Commit**. Repeat as needed.

8. Merge back:  
   Switch back to `master` by clicking the branch name and selecting it. Click **Pull** to ensure it’s up to date. Then use the **branch menu** to merge your feature branch into `master`.

9. Delete the feature branch:  
   Open the branch list, right-click (or use options menu) on your feature branch, and select **Delete Branch**.
