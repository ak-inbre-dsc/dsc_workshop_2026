## &nbsp;&nbsp; 3.1.&nbsp;&nbsp; <span style="color: green;">Collaborator 1</span> ONLY: Create the central Github Repository that you will collaborate on.

**<span style="color: green;">Collaborator 1</span>** will create the central repository on GitHub that both collaborators will contribute to.
<div style="margin-left: 30px;">
    
1.  **Go to GitHub.com:** Log in to your GitHub account.
2.  **Create a New Repository:**
    * Click the `+` sign in the top right corner, then select "New repository."
    * Name your repository: `penguin-collaboration-project`.
    * Make it **Private**. (Public works too, but you may not want everyone to see your practice work.)
    * **IMPORTANT: DO NOT check "Add a README file" or "Add .gitignore" or "Choose a license".** 
    * Click "Create repository."
3.  **Get the Remote URL:** After creating the repository, GitHub will show you a page with instructions. Copy the URL under the header "or push an existing repository from the command line." It will look like `git@github.com:<owner_username>/penguin-collaboration-project.git`. Keep this URL handy!
4.  **Invite Collaborator 2:** On your new GitHub repository page, go to **"Settings"** -> **"Collaborators and teams"**. Click "Add people" and invite your partner (**<span style="color: purple;">Collaborator 2</span>**) by their GitHub username or email. ****<span style="color: purple;">Collaborator 2</span>** must accept this invitation before they can proceed.**
</div>

---
## &nbsp;&nbsp; 3.2.&nbsp;&nbsp; Connect Local Project to New Remote & Push and push your `master` branch.

Before connecting you need to replace the existing 'origin' remote connection to ensure a clean setup. This is now set to your previous repository git@github.com:<Collaborator1_username>/penguins.git. Follow the steps below to 

<div style="margin-left: 30px;">
    
1. **Remove the old remote connection:**
   - Note: The JupyterLab Git Sidebar does not have a native button to remove an existing remote. You must use the Terminal for this specific step.
   - Open a Terminal and type:
</div>

```python
!git remote remove origin
```

<div style="margin-left: 30px;">
    
2. **Add the new shared remote:**
    - In the Git sidebar, click the **Push** icon . Because you removed the previous remote, a dialog box will appear.
    - In the **Remote URL** field, paste the URL for the new shared repository (e.g., `git@github.com:<Collaborator1_username>/penguin-collaboration-project.git`). In the **Remote name** field, type `origin`.
    - Click **Add Remote**.
</div>

<div style="margin-left: 30px;">

3. **Push the master branch:**
    - Ensure the branch dropdown at the top of the sidebar is set to `master`.
    - Click the **Push** icon again. 
    - If prompted to "Publish" or set the upstream branch, confirm the action. This uploads your local `master` branch to the new repository.
</div>

## &nbsp;&nbsp; 3.3.&nbsp;&nbsp; Develop Your Feature Branch

Now that your local project is connected to the shared GitHub repository, **<span style="color: green;">Collaborator 1</span>** will begin developing their feature branch. Change `<Collaborator1_name>` to your actual name here for clarity. You'll create your own branch, make a specific change to `my_code.R`, and then prepare to share it.

Your feature branch will be named `feature/<Collaborator1_name>_plot-title`.


<div style="margin-left: 30px;">

3. Name your new branch `feature/<Collaborator1_name>_plot-title`. You will now automatically be working from the new branch.

## &nbsp;&nbsp; 3.4.&nbsp;&nbsp; Modify `my_code.R` (Add a Plot Title)

Now that you're on your `feature/<Collaborator1_name>_plot-title` branch, you'll modify the `my_code.R` file to add a descriptive title to the plot.

<div style="margin-left: 30px;">

1.  **Locate `my_code.R`:** In the left sidebar of your JupyterLab interface, navigate into the `my_first_git_project` directory. Double-click to open `my_code.R`.
2.  **Add a new line:** At the end of the existing `ggplot` chain (after `theme_minimal()`), add the following line. Make sure to maintain the indentation of the R code:

</div>

---

## &nbsp;&nbsp; 3.5.&nbsp;&nbsp; Commit and Push Your Feature Branch 

Now you will add our changes to your Staging Area, commit it to your local Repository, and push it to the new feature branch on Github. 
<div style="margin-left: 30px;">
    
1. Click the `+` sign that appear when you hover your cursor over `my_code.R`. 
2. Include a concise, descriptive commit message in the Summary text box & Click Commit.
3. Click: **Git → Push to remote** in the Git menu.
</div>

## &nbsp;&nbsp; 3.6.&nbsp;&nbsp; Delete the local feature branch. 
First, ensure you are NOT on the branch you are trying to delete. You should not be, but it is always good to check.
<div style="margin-left: 30px;">
    
1. Open the branch list in the Git sidebar again.
2. Right-click the feature branch (`feature/<Collaborator1_name>_plot-title`).
3. Select **Delete Branch** to keep your workspace organized.

</div>
